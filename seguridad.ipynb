```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np

def de_pca(df_columnas, columnas,numero_PC=1):
  scaler = StandardScaler()
  scaled_data = scaler.fit_transform(df_columnas[columnas])
  pca = PCA(n_components=numero_PC)
  column_reducidas =np.abs(pca.fit_transform(df_columnas[columnas]))
  pesos = pca.components_
  #return pesos
  return column_reducidas, pesos
```

```python
from math import radians, sin, cos, sqrt, atan2

from math import radians, sin, cos, sqrt, atan2

def haversine(point1, point2):
    """
    Calcula la distancia entre dos puntos (lat, lon) usando la fórmula de Haversine.

    Parámetros:
    - point1: Un objeto Point de shapely.
    - point2: Un objeto Point de shapely.

    Retorna:
    - La distancia en kilómetros entre los dos puntos.
    """
    # Radio de la Tierra en kilómetros
    R = 6371.0

    # Extraer latitud y longitud y convertirlas a radianes
    lat1, lon1 = radians(point1.y), radians(point1.x)
    lat2, lon2 = radians(point2.y), radians(point2.x)

    # Diferencias
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # Aplicar la fórmula de Haversine
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # Distancia final en kilómetros
    distance = R * c
    return distance
```

```python
def calcular_distancias_a_todos(centroid, geometries):
    """
    Calcula la distancia entre un punto (centroid) y todos los puntos en una serie de geometrías.

    Parámetros:
    - centroid: El punto de referencia (un objeto Point de shapely).
    - geometries: Una serie de geometrías (points) para calcular las distancias.

    Retorna:
    - Una lista con las distancias en kilómetros entre el centroid y cada geometría.
    """
    distancias = [haversine(centroid, geom) for geom in geometries]
    return distancias



df_seguridad_barrio["distancias_estaciones"] = df_seguridad_barrio.apply(
    lambda row: calcular_distancias_a_todos(row['centroid'], gdf_estaciones["geometry"]), axis=1
)

# Calcular las distancias entre cada centroid y todos los CAIs
df_seguridad_barrio["distancias_cais"] = df_seguridad_barrio.apply(
    lambda row: calcular_distancias_a_todos(row['centroid'], gdf_cai["geometry"]), axis=1
)
# Definir umbral para estaciones cercanas (ej. 1 km)

umbral_cercania = 1

# Para cada barrio, calcular el número de estaciones cercanas
df_seguridad_barrio["estaciones_cercanas"] = df_seguridad_barrio["distancias_estaciones"].apply(
    lambda distancias: sum(np.array(distancias) < umbral_cercania)
)

# Para cada barrio, calcular el número de CAIs cercanos
df_seguridad_barrio["cais_cercanas"] = df_seguridad_barrio["distancias_cais"].apply(
    lambda distancias: sum(np.array(distancias) < umbral_cercania)
)

# Calcular la media de las distancias a todas las estaciones
df_seguridad_barrio["media_distancias_estaciones"] = df_seguridad_barrio["distancias_estaciones"].apply(
    lambda distancias: np.mean(distancias) if len(distancias) > 0 else np.nan
)

# Calcular la media de las distancias a todos los CAIs
df_seguridad_barrio["media_distancias_cais"] = df_seguridad_barrio["distancias_cais"].apply(
    lambda distancias: np.mean(distancias) if len(distancias) > 0 else np.nan
)
```

```python
def calcular_indicador_seguridad(media_distancias, estaciones_cercanas, total_estaciones):
    if media_distancias > 0:  # Asegurar que no haya división por cero
        return (media_distancias) + (estaciones_cercanas)
    else:
        return 0  # O algún valor específico en caso de que no haya estaciones cercanas


df_seguridad_barrio["IndicadorSeguridad_estaciones"] = df_seguridad_barrio.apply(
    lambda row: calcular_indicador_seguridad(row["media_distancias_estaciones"], row["estaciones_cercanas"], len(gdf_estaciones)),
    axis=1
)

df_seguridad_barrio["IndicadorSeguridad_cais"] = df_seguridad_barrio.apply(
    lambda row: calcular_indicador_seguridad(row["media_distancias_cais"], row["cais_cercanas"], len(gdf_estaciones)),
    axis=1
)
```

```python
import pandas as pd

# Suponiendo que 'cama' es tu DataFrame y ya tiene la columna 'TasaSeguridadInseguridad'
min_rango = 1
max_rango = 5

# Identificar outliers
Q1 = cama['TasaSeguridadInseguridad'].quantile(0.25)
Q3 = cama['TasaSeguridadInseguridad'].quantile(0.75)
IQR = Q3 - Q1

# Umbrales para identificar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Crear una nueva columna para calificación
cama['tasacalificada'] = None

# Normalizar valores dentro del rango
normal_values = cama[(cama['TasaSeguridadInseguridad'] >= lower_bound) & (cama['TasaSeguridadInseguridad'] <= upper_bound)]
min_columna = normal_values['TasaSeguridadInseguridad'].min()
max_columna = normal_values['TasaSeguridadInseguridad'].max()

# Aplicar la transformación para los valores normales
cama.loc[normal_values.index, 'tasacalificada'] = ((normal_values['TasaSeguridadInseguridad'] - min_columna) / (max_columna - min_columna)) * (max_rango - min_rango) + min_rango

# Asignar calificaciones para outliers
# Para outliers superiores
outliers_superiores = cama[cama['TasaSeguridadInseguridad'] > upper_bound]
if not outliers_superiores.empty:
    outliers_superiores['tasacalificada'] = 5 + ((outliers_superiores['TasaSeguridadInseguridad'] - upper_bound) / (cama['TasaSeguridadInseguridad'].max() - upper_bound) * 1)  # Normalización a partir de 5
    cama.loc[outliers_superiores.index, 'tasacalificada'] = outliers_superiores['tasacalificada']

# Para outliers inferiores
outliers_inferiores = cama[cama['TasaSeguridadInseguridad'] < lower_bound]
if not outliers_inferiores.empty:
    cama.loc[outliers_inferiores.index, 'tasacalificada'] = 1  # Calificación de 1

# Imprimir resultados
print(cama[['TasaSeguridadInseguridad', 'tasacalificada']])
```
